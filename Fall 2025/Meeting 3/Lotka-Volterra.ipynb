# Lotka-Volterra Equations

## Import Statements

```python
import numpy as np
from matplotlib import pyplot as plt
from my_integrators_solution import ode_integrator
```

## Introducing the System of ODEs

$
\begin{align*}
\frac{\mathrm{d} N}{\mathrm{d} t} &= a N - b N P \\
\frac{\mathrm{d} P}{\mathrm{d} t} &= c N P - d P
\end{align*}
$

```python
# Define Lotka-Volterra Equations
a = 1
b = -1
c = 1
d = -1

f = lambda t, y: np.array([
    a * y[0] + b * y[0] * y[1],
    c * y[0] * y[1] + d * y[1]
])
```

```python
# Solve ODE
t0 = 0
tf = 20
dt = 0.01
x0 = np.array([2, 1])
ts, ys = ode_integrator('explicit_rk4', f, (t0, tf), x0, dt)
```

```python
ys
```

```python
fig, ax = plt.subplots()
ax.plot(ts, [y[0] for y in ys], label='prey')
ax.plot(ts, [y[1] for y in ys], label='predator')
```
